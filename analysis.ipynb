```python
# data in size column contains with MB's and Kb's in string format, now we should convert it to int format datatype
df.dtypes
```

```python
# Variables seem to have incorrect type and inconsistent formatting. You need to fix them:
# Size column has sizes in Kb as well as Mb.
# To analyze, you’ll need to convert these to numeric.
# Extract the numeric value from the column and Multiply the value by 1,000, if size is mentioned in Mb

# now we will write a programme using function to convert Size column to integer

def size_col_processing(x):
    x=str(x.lower())
    if 'm' in x:
        
        val=float(x.replace('m', ''))
        val=val*1000
        
    elif 'k' in x:
        val=float(x.replace('k', ''))
        
        
    else:
        val=0
    return val
```

```python
# we need to APPLY function to that specific column
df['Size']=df['Size'].apply(size_col_processing)  
```

```python
df  
df.dtypes
```

```python
# Reviews is a numeric field that is loaded as a string field. Convert it to numeric (int/float).

df['Reviews']=df['Reviews'].astype(float)
```

```python
df.dtypes
```

```python
# Installs field is currently stored as string and has values like 1,000,000+. 
# Treat 1,000,000+ as 1,000,000
# remove ‘+’, ‘,’ from the field, convert it to integer

def Installs_col_processing(x):
    x=x.replace('+','')
    x=x.replace(',','')
    x=float(x)
    return x
```

```python
df['Installs']=df['Installs'].apply(Installs_col_processing)
```

```python
df.head()
```

```python
# Price field is a string and has $ symbol. Remove ‘$’ sign, and convert it to numeric.

df['Price']=df['Price'].apply(lambda x:float(str(x).replace('$','')))
```
